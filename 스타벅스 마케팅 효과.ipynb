```python
import pandas as pd

def calc_offer_metrics(df):
    results = []

    for offer_type in df['offer_type'].unique():
        data = df[df['offer_type'] == offer_type]

        # 1. 받은 사람 (offer received)
        received = (
            data[data['event'] == 'offer received']
            .groupby(['person', 'offer_id'])
            .size()
            .reset_index()[['person', 'offer_id']]
        )
        n_received = len(received)

        # 2. 완료한 사람 (offer completed)
        completed = (
            data[data['event'] == 'offer completed']
            .groupby(['person', 'offer_id'])
            .size()
            .reset_index()[['person', 'offer_id']]
        )

        # 전환율 계산 (offer received 대비 offer completed)
        n_completed = len(pd.merge(received, completed, on=['person', 'offer_id'], how='inner'))
        conversion_rate = round(n_completed / n_received, 3) if n_received > 0 else 0

        # 재참여율: 구매(offer completed) 2회 이상인 고객 비율
        purchase_counts = (
            data[data['event'] == 'offer completed']
            .groupby('person')
            .size()
        )
        loyal_customers = (purchase_counts >= 2).sum()
        total_customers = purchase_counts.shape[0]
        repurchase_rate = round(loyal_customers / total_customers, 3) if total_customers > 0 else 0

        # 응답시간: offer received ~ offer completed 시간 차이 평균
        received_time = (
            data[data['event'] == 'offer received']
            .groupby(['person', 'offer_id'])['time']
            .min()
            .reset_index(name='received_time')
        )
        completed_time = (
            data[data['event'] == 'offer completed']
            .groupby(['person', 'offer_id'])['time']
            .min()
            .reset_index(name='completed_time')
        )
        merged_time = pd.merge(received_time, completed_time, on=['person', 'offer_id'], how='inner')
        merged_time['duration'] = merged_time['completed_time'] - merged_time['received_time']
        avg_response_time = merged_time['duration'].mean() if not merged_time.empty else None

        results.append({
            'offer_type': offer_type,
            'conversion_rate': conversion_rate,
            'repurchase_rate': repurchase_rate,
            'avg_response_time': avg_response_time
        })

    return pd.DataFrame(results).set_index('offer_type')
```
